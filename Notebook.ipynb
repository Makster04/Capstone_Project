```python
# Load the uploaded CSV files
city_df = pd.read_csv("Edited_Business_City_df.csv")
country_df = pd.read_csv("Edited_Business_Country_df.csv")

# Group cities by country and join them into a single string

# Merge the grouped cities into the country-level dataframe
combined_df = pd.concat([country_df, city_df], ignore_index=True)



# Display the updated dataframe to the user

combined_df_sorted = combined_df.sort_values(by="Country").reset_index(drop=True)

combined_df_sorted["City"] = combined_df_sorted.apply(
    lambda row: f"{row['City'].strip(', ')}" + ", " + f"{row['Country'].strip(', ')}" 
    if pd.notnull(row["City"]) and pd.notnull(row["Country"]) else row["City"],
    axis=1)


combined_df_sorted
```

```python
import pandas as pd
import sqlite3

# Load CSV files into DataFrames
Cost_of_Living_df = pd.read_excel('Stats/NUMBEO_Cost_of_Living_Index.xlsx')
Crime_and_Safety_df = pd.read_excel('Stats/NUMBEO_Crime_and_Safety_Index.xlsx')
Healthcare_df = pd.read_excel('Stats/NUMBEO_Healthcare_Index.xlsx')
Pollution_df = pd.read_excel('Stats/NUMBEO_Pollution_Index.xlsx')
Property_Price_df = pd.read_excel('Stats/NUMBEO_Property_Price_Index.xlsx')
Quality_of_life_df = pd.read_excel('Stats/NUMBEO_Quality_of_Life_Index.xlsx')
Traffic_df = pd.read_excel('Stats/NUMBEO_Traffic_Index.xlsx')

conn = sqlite3.connect(':memory:')

# Load DataFrames into the database as tables
Cost_of_Living_df.to_sql('CostOfLiving', conn, index=False, if_exists='replace')
Crime_and_Safety_df.to_sql('CrimeSafety', conn, index=False, if_exists='replace')
Healthcare_df.to_sql('Healthcare', conn, index=False, if_exists='replace')
Pollution_df.to_sql('Pollution', conn, index=False, if_exists='replace')
Property_Price_df.to_sql('PropertyPrice', conn, index=False, if_exists='replace')
Quality_of_life_df.to_sql('QualityOfLife', conn, index=False, if_exists='replace')
Traffic_df.to_sql('Traffic', conn, index=False, if_exists='replace')



# SQL query to merge all tables on 'Player'
sql_query = """
SELECT 
    PropertyPrice.City,
    PropertyPrice.[Price To Income Ratio],
    PropertyPrice.[Gross Rental Yield City Centre],
    PropertyPrice.[Gross Rental Yield Outside of Centre],
    PropertyPrice.[Price To Rent Ratio City Centre],
    PropertyPrice.[Price To Rent Ratio Outside Of City Centre],
    PropertyPrice.[Mortgage As A Percentage Of Income],
    PropertyPrice.[Affordability Index],

    QualityOfLife.[Quality of Life Index],
    QualityOfLife.[Purchasing Power Index],
    QualityOfLife.[Safety Index],
    QualityOfLife.[Health Care Index],
    QualityOfLife.[Cost of Living Index],
    QualityOfLife.[Property Price to Income Ratio],
    QualityOfLife.[Traffic Commute Time Index],
    QualityOfLife.[Pollution Index],
    QualityOfLife.[Climate Index],

    Traffic.[Traffic Index],
    Traffic.[Time Index],
    Traffic.[Time Exp. Index],
    Traffic.[Inefficiency Index],
    Traffic.[CO2 Emission Index],

    Pollution.[Pollution Index] AS PollutionIndex_Detail,
    Pollution.[Exp Pollution Index],

    Healthcare.[Health Care Index] AS HealthcareIndex_Detail,
    Healthcare.[Health Care Expenditure Index],

    CrimeSafety.[Crime Index],
    CrimeSafety.[Safety Index] AS SafetyIndex_Crime,

    CostOfLiving.[Cost of Living Index] AS CostOfLivingIndex_Detail,
    CostOfLiving.[Rent Index],
    CostOfLiving.[Cost of Living Plus Rent Index],
    CostOfLiving.[Groceries Index],
    CostOfLiving.[Restaurant Price Index],
    CostOfLiving.[Local Purchasing Power Index]

FROM PropertyPrice
JOIN QualityOfLife ON PropertyPrice.City = QualityOfLife.City
JOIN Traffic ON PropertyPrice.City = Traffic.City
JOIN Pollution ON PropertyPrice.City = Pollution.City
JOIN Healthcare ON PropertyPrice.City = Healthcare.City
JOIN CrimeSafety ON PropertyPrice.City = CrimeSafety.City
JOIN CostOfLiving ON PropertyPrice.City = CostOfLiving.City
WHERE PropertyPrice.City LIKE '%United States%'



"""




# Execute the query and load results into a DataFrame
All_Numbeo_Numbers_df = pd.read_sql_query(sql_query, conn)

# Save the merged DataFrame to a new CSV file
csv_filename = "Stats/All_Numbeo_Numbers.csv"
All_Numbeo_Numbers_df.to_csv(csv_filename, index=False)

# Close the connection
conn.close()

# Provide the file to the user
All_Numbeo_Numbers_df.colums
```
